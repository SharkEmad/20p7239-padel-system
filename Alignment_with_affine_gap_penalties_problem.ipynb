#Alignment with affine gap penalties problem

```python
def HighestScoringAlignment(v, w, ScoreMatrix, GOP , GEP):
    n = len(v)
    m = len(w)

    #Initialize Score Matrix
    S = []
    for i in range(n+1):
      row = []
      for j in range(m+1):
         row.append(0)
      S.append(row)

    #Initialize Horizontal Gap Matrix
    Horizontal = []
    for i in range(n+1):
        row = []
        for j in range(m+1):
            row.append(float('-inf'))
        Horizontal.append(row)

    #Initialize Vertical Gap Matrix
    Vertical = []
    for i in range(n+1):
        row = []
        for j in range(m+1):
            row.append(float('-inf'))
        Vertical.append(row)

    # Initialize the traceback matrix
    TracebackMatrix = []
    for i in range(n+1):
        row = []
        for j in range(m+1):
            row.append('-')
        TracebackMatrix.append(row)


    # Initialize first row of both Horizontal Gap Matrix and Score Matrix
    for i in range(1, n+1):
        Horizontal[i][0] = GOP + (i-1) * GEP
        S[i][0] = float('-inf')
        TracebackMatrix[i][0] = 'U'

    # Initialize first column of both Vertical Gap Matrix and Score Matrix
    for j in range(1, m+1):
        Vertical[0][j] = GOP + (j-1) * GEP
        S[0][j] = float('-inf')
        TracebackMatrix[0][j] = 'L'

    # Fill in the matrices
    for i in range(1, n+1):
        for j in range(1, m+1):
            S[i][j] = max(S[i-1][j-1] + ScoreMatrix[v[i-1]][w[j-1]],          # Matrix + Score(Match,Mismatch)
                          Horizontal[i-1][j-1] + ScoreMatrix[v[i-1]][w[j-1]], # Ix + Score(Match,Mismatch)
                          Vertical[i-1][j-1] + ScoreMatrix[v[i-1]][w[j-1]])   # Iy + Score(Match,Mismatch)

            Horizontal[i][j] = max(S[i-1][j] + GOP + GEP,      # Matrix + Gap Open + Gap Extension
                                   Horizontal[i-1][j] + GEP)   # Matrix + Gap Extension

            Vertical[i][j] = max(S[i][j-1] + GOP + GEP,     # Matrix + Gap Open + Gap Extension
                                 Vertical[i][j-1] + GEP)    # Matrix + Gap Extension

            if S[i][j] == max(S[i][j],Horizontal[i][j],Vertical[i][j]): #Check if the new value came from the diagonal
                TracebackMatrix[i][j] = 'D' #This is for Diagonal traceback
            elif Horizontal[i][j] == max(S[i][j],Horizontal[i][j],Vertical[i][j]): #Check if the new value came from the Upper
                TracebackMatrix[i][j] = 'U' #This is for Up traceback
            else: #Check if the new value came from the Left
                TracebackMatrix[i][j] = 'L' #This is for Left traceback

    # Traceback
    align1 = []
    align2 = []

    i, j = n, m
    while i > 0 or j > 0:
        if i > 0 and j > 0 and TracebackMatrix[i][j] == 'D':
            align1.append(v[i-1])
            align2.append(w[j-1])
            i -= 1
            j -= 1
        elif i > 0 and TracebackMatrix[i][j] == 'U':
            align1.append(v[i-1])
            align2.append('-')
            i -= 1
        else:
            align1.append('-')
            align2.append(w[j-1])
            j -= 1


    align1 = ''.join(reversed(align1))
    align2 = ''.join(reversed(align2))

    return align1, align2, S[n][m] , S , Horizontal , Vertical , TracebackMatrix
```
